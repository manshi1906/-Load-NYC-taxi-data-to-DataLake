Load NYC taxi data to DataLake/Blob_Storage/DataBricks and extract the data through dataframe in the notebook.
Perform Following Queries using PySpark.

Query 1. - Add a column named as "Revenue" into dataframe which is the sum of the below columns 'Fare_amount','Extra','MTA_tax','Improvement_surcharge','Tip_amount','Tolls_amount','Total_amount'

Query 2. - Increasing count of total passengers in New York City by area

Query 3. - Realtime Average fare/total earning amount earned by 2 vendors

Query 4. - Moving Count of payments made by each payment mode

Query 5. - Highest two gaining vendor's on a particular date with no of passenger and total distance by cab

Query 6. - Most no of passenger between a route of two location.

Query 7. - Get top pickup locations with most passengers in last 5/10 seconds.

```python
from  pyspark.sql.functions import *
path='dbfs:/FileStore/shared_uploads/rahulnyati8@gmail.com/yellow_tripdata_2018_01.parquet'
module_name = "spark-programming"
spark.conf.set("com.databricks.training.module-name", module_name)
data=spark.read.parquet(path)
```

#####Printing Schema

```python
data.printSchema()
```

```python
display(data)
```

#####Query 1. - Add a column named as ""Revenue"" into dataframe which is the sum of the below columns 'Fare_amount','Extra','MTA_tax','Improvement_surcharge','Tip_amount','Tolls_amount','Total_amount'

```python
new_data1=(data.withColumn("Revenue",(col("Fare_amount"))+(col("Extra"))+(col("MTA_tax"))+(col('Improvement_surcharge'))+(col('Tip_amount'))+(col('Tolls_amount'))+(col('Total_amount'))))
```

```python
display(new_data1)
```

#####Query 2. - Increasing count of total passengers in New York City by area

```python
increasing_count=new_data1.groupBy("PULocationID","DOLocationID").agg(sum("passenger_count").alias("Pass_Count"))
```

```python
increasing_count.printSchema()
```

```python
#Increasing Count of Total Passengers
display(increasing_count)
```

#####Query 3. -Realtime Average fare/total earning amount earned by 2 vendors

```python
Realtime_avg=new_data1.groupBy("VendorID").agg(avg("fare_amount").alias("Fare_Amount_Avg"),avg("Total_amount").alias("Total_Amount_avg"))
```

```python
display(Realtime_avg)
```

#####Query 4. - Moving Count of payments made by each payment mode

```python
Moving_count=new_data1.groupBy("payment_type").agg(count("payment_type").alias("Count_of_Payment"))
```

```python
#1= Credit card
#2= Cash
#3= No charge
#4= Dispute
#5= Unknown
#6= Voided trip
display(Moving_count)
```

#####Query 5. -Highest two gaining vendor's on a particular date with no of passenger and total distance by cab

```python
Higest_gaining_vendor=new_data1.groupBy("VendorID","tpep_pickup_datetime").agg(sum("passenger_count").alias("No of Passenger"),sum("trip_distance").alias("Total_Distance"))
```

```python
display(Higest_gaining_vendor)
```

#####Query 6. -Most no of passenger between a route of two location.

```python
Max_no_of_pass=new_data1.groupBy("PULocationID","DOLocationID").agg(max("passenger_count"))
```

```python
display(Max_no_of_pass)
```

#####Query 7. - Get top pickup locations with most passengers in last #5/10 seconds."

```python
Most_Passenger=new_data1.groupBy("PULocationID",window(col("tpep_pickup_datetime"),"5 seconds").alias("Wd_value")).agg(max("passenger_count"))
```

```python
import pyspark.sql.functions as F
Most_Passenger.sort(F.col("Wd_value.start").desc()).limit(1).show()
```
